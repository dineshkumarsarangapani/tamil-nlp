## Script to extract ponniyan selvan book

from: http://book.ponniyinselvan.in/part-1/chapter-1.html

Each part is divided as chaps
* part 1 : 57 chaps
* part 2: 53 Chaps
* part 3: 46 Chaps
* part 4: 46 Chaps
* part 5: 91 chaps

```python
import urllib.request
from bs4 import BeautifulSoup

user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
```

```python
location = {1:57,2:53,3:46,4:46,5:92}
```

```python
def get_data(url):
    headers={'User-Agent':user_agent,} 
    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
    response.close()
    data_utf8 = data.decode("utf-8") 
    soup = BeautifulSoup(data_utf8)
    tag=soup.find("section", {"class": "markdown-section"})
    return tag.get_text()
```

```python
def append_file(data):
    with open('ponniayan.txt', 'a') as fd:
        fd.write(f'\n{data}')
```

```python
for part in location:
    for chap in range(1,location[part]):
        url = "http://book.ponniyinselvan.in/part-"+str(part)+"/chapter-"+str(chap)+".html"
        print(url)
        data = get_data(url)
        print(data[1:200])
        append_file(data)
```
